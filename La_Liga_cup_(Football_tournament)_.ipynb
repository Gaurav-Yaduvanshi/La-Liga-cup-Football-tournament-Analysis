# A function with name Goal_diff_count which should return all the teams with their Goal Differences

```python
def Goal_diff_count(df):
  df['Goal_diff_count'] = df['GoalsFor'].astype(int) - df['GoalsAgainst'].astype(int)
  return df[['Team','Goal_diff_count']].sort_values(by = 'Goal_diff_count',ascending=False)
```

```python
Goal_diff_count(df)
```

# Find the team which has maximum and minimum goal difference.

```python
# Team which has the maximum goal difference
Goal_diff_count(df).head(1)
```

```python
# Team which has the minimum goal difference
Goal_diff_count(df).tail(1)
```

##  Create a new column with name Winning Percent and append it to the data set. If there are any numerical error, replace it with 0%

```python
#Percentage of Winning = (GamesWon / GamesPlayed)*100

df['Winning Percent'] = (df['GamesWon'].astype(int)/df['GamesPlayed'].astype(int))*100

df['Winning Percent'].fillna(0,inplace = True)
```

```python
# the top 5 teams which has the highest Winning percentage

df[['Team','Goal_diff_count','Winning Percent']].sort_values(by = 'Winning Percent',ascending = False).head(5)
```
